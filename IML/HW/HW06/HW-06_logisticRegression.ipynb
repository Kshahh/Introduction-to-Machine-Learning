2. Logistic regression: Classify individuals into 3 categories, as in HW5.


```python
# Based on the ratio (citations in 2022)/(citations in 2021),
# approximated to two decimal places, determine the category of each individual as one of the three shown below:
# 1. Low (<1.05).
# 2. Medium (1.06-1.15).
# 3. High (>1.15).
```

```python
# Import Modules 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LogisticRegression
```

```python
# Import dataset

dataset = pd.read_csv('61-70.csv')
```

```python
df = dataset.copy()
```

```python
df.head()
```

```python
# Stastical analysis of df

df.info()
```

```python
# We need to change the cit_2017 Dtype from Object to int64

df['cit_2017'] = df['cit_2017'].str.replace(",","").astype(int)
```

```python
df.info()
```

```python
# Calculate the (citations in 2022) / (citations in 2021) ratio for each individual. 
cit_ratio = df['cit_2022'] / df['cit_2021']
cit_ratio = cit_ratio.round(2)
df['citation_ratio'] = cit_ratio
```

```python
# 1. Low (<1.05).
low = df['citation_ratio'] < 1.05
```

```python
# 2. Medium (1.06-1.15).
# lol_medium = df['citation_ratio'] > 1.06 & df['citation_ratio'] < 1.15
medium = (df['citation_ratio'] >= 1.06) & (df['citation_ratio'] <= 1.15)
```

```python
# 3. High (>1.15).
high = df['citation_ratio'] > 1.15
```

```python
# Determine the category of each individual as one of the three shown below:
# 1. Low (<1.05). == 0
# 2. Medium (1.06-1.15). == 1
# 3. High (>1.15). == 2
```

```python
df.loc[df['citation_ratio'] <= 1.05, 'cit_category'] = "0"
df.loc[(df['citation_ratio'] >= 1.06) & (df['citation_ratio'] <= 1.15), 'cit_category'] = "1"
df.loc[df['citation_ratio'] > 1.15, 'cit_category' ] = "2"
```

```python
df.head()
```

```python
#  Import train_test_slit
from sklearn.model_selection import train_test_split

# Splitting the dataset into 80% training and 20% testing 

x = df[['cit_2017', 'cit_2018', 'cit_2019', 'cit_2020', 'cit_2021','cit_2022']]
y = df[['cit_category']]

# Splitting your DataFrame 'df' into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size = 0.80, test_size=0.2, random_state=42)
```

```python
# # Normalizing both the train and test data

# from sklearn.preprocessing import MinMaxScaler
# scaler = MinMaxScaler(feature_range=(0, 1))

# x_train = scaler.fit_transform(x_train)


# x_test = scaler.transform(x_test)
```

```python
# Implying Logistic Regression:

model_logreg = LogisticRegression()
model_logreg.fit(x_train,y_train)
y_pred = model_logreg.predict(x_test)
```

```python
# calculating MSE
mse = mean_squared_error(y_test,y_pred)
print(f'Logistic Regression Mean Squared Error: {mse}')
```

```python
# Convert arrays to DataFrames for better visualization
df_x_test = pd.DataFrame(x_test, columns=['cit_2017', 'cit_2018', 'cit_2019', 'cit_2020', 'cit_2021','cit_2022'])
df_y_test = pd.DataFrame(y_test, columns=['cit_category'])
df_y_pred = pd.DataFrame(y_pred, columns=['Predicted_cit_category'])

# Reset indices before concatenating
df_x_test.reset_index(drop=True, inplace=True)
df_y_test.reset_index(drop=True, inplace=True)
df_y_pred.reset_index(drop=True, inplace=True)

# Concatenate the DataFrames horizontally
df_results = pd.concat([df_x_test, df_y_test, df_y_pred], axis=1)
df_results.to_csv('Logistic_regression_predicted_values.csv', index=False)
# Display the DataFrame with x_test, y_test, and predicted values
print(df_results)
```
