```python
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Upload Data Set

dataset = pd.read_csv('61-70.csv')
dataset.head()
```

```python
# Copy dataset into data frame
df = dataset.copy()
df.head()
```

```python
# Currently the cit_2017 Dtype is object hence converting it to Dtype int64 

df['cit_2017'] = df['cit_2017'].str.replace(',', "").astype(int)
```

```python
# Splitting the dataset into 80% training and 20% testing 

x = df[['cit_2017', 'cit_2018', 'cit_2019', 'cit_2020', 'cit_2021']]
y = df[['cit_2022']]

# Splitting your DataFrame 'df' into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size = 0.80, test_size=0.2, random_state=42)
```

```python
# Normalizing both the train and test data

from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
x_train = scaler.fit_transform(x_train)
x_train
```

```python
x_test = scaler.transform(x_test)
x_test
```

```python
# Importing tensorflow

import tensorflow as tf
from random import seed
from random import random

lol = len(x_train[0])
lol
```

```python
# Defining the neural network 5-3-1 architecture using Tensor Flow:

seed(1)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(3, activation = 'relu' , input_shape = (len(x_train[0]),)),
    tf.keras.layers.Dense(1, activation = 'relu')
    
])

model.summary()
```

```python
# defining hyper parameters:
learning_rate = 0.1
model.compile( tf.keras.optimizers.SGD(learning_rate), loss = 'mean_absolute_error')
model.fit(x_train, y_train, epochs = 1000 , shuffle = False)
    
```

```python
# Calculate MAE

from sklearn.metrics import mean_absolute_error
y_pred = model.predict(x_test)
mean_absolute_error = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mean_absolute_error}")
```
