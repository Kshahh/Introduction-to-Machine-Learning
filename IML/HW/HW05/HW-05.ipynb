```python
#  Import train_test_slit
from sklearn.model_selection import train_test_split
```

```python
# Splitting the dataset into 80% training and 20% testing 

x = df[['cit_2017', 'cit_2018', 'cit_2019', 'cit_2020', 'cit_2021','cit_2022']]
y = df[['Low','Medium','High']]

# Splitting your DataFrame 'df' into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size = 0.80, test_size=0.2, random_state=42)
```

```python
# Normalizing both the train and test data

from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
x_train = scaler.fit_transform(x_train)
x_train
```

```python
x_test = scaler.transform(x_test)
x_test
```

```python
# Importing tensorflow

import tensorflow as tf
from random import seed
from random import random

lol = len(x_train[0])
lol
```

```python
# Defining the neural network 6-6-3 architecture using Tensor Flow:

seed(1)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(6, activation = 'tanh' , input_shape = (len(x_train[0]),)),
    tf.keras.layers.Dense(3, activation = 'sigmoid')
])

model.summary()
```

```python
# defining hyper parameters:
learning_rate = 0.1
model.compile( tf.keras.optimizers.SGD(learning_rate), loss = 'categorical_crossentropy', metrics = ['accuracy'])
model.fit(x_train, y_train, epochs = 3800 , shuffle = False)
    
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print("Accuracy: " , accuracy)
print("Loss: ", loss)
```

```python
y_pred = model.predict(x_test)
y_pred
```
